Tensors in Pytorch

```python
# In Pytorch it's not array anymore, so dim is used to describe about it's shape!
x_dim0 = torch.tensor(2)
x_dim1 = torch.tensor([1, 2, 3])
x_dim2 = torch.tensor([[1,2,3]])
x_dim3 = torch.tensor([[[1,2,3]]])
print(x_dim0.dim()) # ~ x.ndim in numpy
print(x_dim1.dim())
print(x_dim2.dim())
print(x_dim3.dim())
```

```python
# In spite of the reason above, It's still likely to NumPy in use!
# But we should call it tensor instead of array!
A = torch.randint(0,255, (2,2)) # ~ np.random.randint(0, 255, (2, 2))
print(type(A))
print(A)
print(A.shape) # can also use view!
print(A.dim())
```

Operator and Linear Algebra extension

```python
A = torch.tensor([[1,2],[3,4]])
B = torch.tensor([[2,3],[4,5]])

# Basic
AaddB = A+B
print('>>> A + B =\n', AaddB)
AminB = A-B
print('>>> A - B =\n', AminB)
Amul2 = A*2
print('>>> A * 2 =\n', Amul2)
AmulB = A*B
print('>>> Hamadard Product: A * B =\n', AmulB)
AdivB = A/B
print('>>> Elements wise: A / B =\n', AdivB)
Apow2 = A**2
print('>>> A**2 =\n', Apow2)
_2powA = 2**A
print('>>> 2**A =\n', _2powA)
Asin = torch.sin(A) # affect a function!
print('>>> sin(A) =\n', Asin)

# Linear Algebra
AB = torch.mm(A,B)# AB = torch.dot(A,B) and A.dot(B) - 1D arrays only!!! To anoying!
print('>>> AB =\n',AB)
Ainv = torch.linalg.pinv(A.type(torch.float32)) # Too anoying!!!!!
print('>>> A^-1 =\n', Ainv)
print('>>> A.T =\n', A.T)
Adet = torch.linalg.det(A.type(torch.float32))
print('>>> detA =\n', Adet)
print('>>> traceA =\n', A.trace())
Afro_norm = torch.linalg.norm(A.type(torch.float32), ord = 'fro')
print('>>> A forbenious norm =\n', Afro_norm)

# ...
```

```python
# So with linear algebra, it's so anoying! Because before some opertors we have to change the type of tensor!
```

Exercise: Softmax_stable

```python
def Softmax_stable(Z):
  return torch.exp(Z-torch.max(Z, dim =1, keepdim = True)[0])/torch.sum(torch.exp(Z-torch.max(Z, dim =1, keepdim = True)[0]), dim = 1, keepdim = True)
```

```python
Z = torch.tensor([[1,2,1,3,5,1],
                  [4,5,2,1,2,4],
                  [1,3,6,7,3,7]])
A = Softmax_stable(Z)
print(A)
```

```python
print(torch.sum(A, dim = 1, keepdim = True))
```
